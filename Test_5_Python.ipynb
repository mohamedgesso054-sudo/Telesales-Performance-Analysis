```python
top_sellers = df_excel.groupby('Name')['Deal Value ($)'].sum().sort_values(ascending = False)
top_sellers
```

```python
most_active = df_excel.groupby('Name')['Total Calls'].sum().sort_values(ascending = False)
print("إجمالي مكالمات الموظفين:")
most_active
```

```python
conversation_rate = df_excel.groupby('Name')['Call Conversion Rate (%)'].sum().sort_values(ascending=False)
print("متوسط نسبة تحويل المكالمات (%) لكل موظف:")
conversation_rate
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_sellers.index,
            y=top_sellers.values,
            hue=top_sellers.index,
            palette='viridis',
            legend=False
            )

plt.title('Total Employee Sales ($)', fontsize=16)

plt.title('Total Employee Sales ($)', fontsize=16)
plt.xlabel('Employee Name', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.xticks(rotation=45)
plt.savefig('sales_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
